```python
# Criar o cardápio
print('Bem vindo a lanchonete do Júlio Henrique Passo de Araújo\n')
print('*' *  20, 'CARDÁPIO', '*' * 20)
print('\n| Código |         Descrição         |   Valor  |')
print('|   100  |      Cachorro Quente      | R$  9,00 |')
print('|   101  |   Cachorro-Quente Duplo   | R$ 11,00 |')
print('|   102  |           X-Egg           | R$ 12,00 |')
print('|   103  |          X-Salada         | R$ 13,00 |')
print('|   104  |          X-Bacon          | R$ 14,00 |')
print('|   105  |          X-Tudo           | R$ 17,00 |')
print('|   200  |     Refrigerante Lata     | R$  5,00 |')
print('|   201  |        Chá Gelado         | R$  4,00 |')
print('')

# acumulador para contar a valor total dos lanches. 
acumulador = 0
#while True sera utilizado para continuar pedindo mais lanches.
while True:
  codigo = int(input('Entre com o código desejado: '))
  #verificaremos se o código inserido é valido.
  if codigo != 100 and codigo != 101 and codigo != 102 and codigo != 103 and codigo != 104 and codigo != 105 and codigo != 200 and codigo != 201:
    print('Opção inválido!')
    continue #Se for inválido o continue fará retornar a pergundo do código do lanche.

  
  elif codigo == 100:
    print('Você escolheu Cachorro Quente no valor de R$9,00')
    acumulador += 9 #pegue o valor que tinha no acumulador e some com 9
  
  elif codigo == 101:
    print('Você escolheu Cachorro-Quente Duplo no valor de R$11,00')
    acumulador += 11 #pegue o valor que tinha no acumulador e some com 9
  
  elif codigo == 102:
    print('Você escolheu X-Egg no valor de R$12,00')
    acumulador += 12 #pegue o valor que tinha no acumulador e some com 9

  elif codigo == 103:
    print('Você escolheu X-Salada no valor de R$13,00')
    acumulador += 13 #pegue o valor que tinha no acumulador e some com 9

  elif codigo == 104:
    print('Você escolheu X-Bacon no valor de R$14,00')
    acumulador += 14 #pegue o valor que tinha no acumulador e some com 9

  elif codigo == 105:
    print('Você escolheu X-Tudo no valor de R$17,00')
    acumulador += 17 #pegue o valor que tinha no acumulador e some com 9

  elif codigo == 200:
    print('Você escolheu Refrigerante Lata no valor de R$5,00')
    acumulador += 5 #pegue o valor que tinha no acumulador e some com 9

  elif codigo == 201:
    print('Você escolheu Chá Gelado no valor de R$4,00')
    acumulador += 4 #pegue o valor que tinha no acumulador e some com 9

  #variável para perguntar se tem mais pedidos
  pedir_maisRU4133412 = int(input('Deseja pedir mais alguma coisa? \n1 - Sim\n0 - Não\n'))
  if pedir_maisRU4133412 == 1:
    continue #Se tiver o continue vai voltar para a entrada de código dos lanches
  elif pedir_maisRU4133412 == 0:
    print('O valor totala ser pago: R${:.2f}'.format(acumulador))
    break #Caso não tenha mais pedidos sera mostrado o valor total e o programa encerra com break
```

```python
#Inicio da Função dimensoesObjeto.
def dimensoesObjeto():
  print('-------------------- Menu 1 de 3 - dimensões do objeto --------------------')
  while True:
    #usaremos try e except caso o usuário digite um não numeral.
    try:
      alturaObjeto = float(input('Digite a altura do objeto (em cm): '))
      larguraObjeto = float(input('Digite a largura do objeto (em cm): '))
      comprimentoObjeto = float(input('Digite o comprimento do objeto (em cm): '))
      #Operação para determinar o volume.
      volumeObjeto = alturaObjeto * larguraObjeto * comprimentoObjeto
      valorObjeto = 0
      #condições para definir o valor final do objeto baseado no volume.
      if volumeObjeto < 1000:
        valorObjeto = 10
      elif 1000 <= volumeObjeto < 10000:
        valorObjeto = 20
      elif 10000 <= volumeObjeto < 30000:
        valorObjeto = 30
      elif 30000 <= volumeObjeto < 100000:
        valorObjeto = 50
      elif volumeObjeto >= 100000:
        print('O volume do objeto em cm³ é: {:.2f}'.format(volumeObjeto))
        print('Volume não aceito! O volume precisa ser menor que 100.000cm³')
        continue
      return valorObjeto, volumeObjeto
    except ValueError:
      print('Você digitou um valor não numérico!\n'
            'Por favor entre com as dimensões novamente')
#Fim da Função dimensoesObjeto

#Inicio da Função pesoObjeto
def pesoObjeto():
  print('-------------------- Menu 2 de 3 - peso do objeto --------------------')
  while True:
    #usaremos try e except caso o usuário digite um não numeral.
    try:
      pesoObjeto = float(input('Digite o peso do objeto (em Kg): '))
      if pesoObjeto <= 0.1:
        return 1
      if 0.1 <= pesoObjeto < 1:
        return 1.5
      elif 1 <= pesoObjeto < 10:
        return 2
      elif 10 <= pesoObjeto < 30:
        return 3
      elif pesoObjeto >= 30:
        print('Não aceitamos objetos com esse peso!\n'
            'Por favor entre com peso novamente')
        continue
      return pesoObjeto
    except ValueError:
      print('Você digitou o peso do obejeto com valor não numérico!\n'
            'Por favor digite o peso em Kg novamente')
#Fim da Função pesoObjeto

#Inicio da Função rotaObjeto
def rotaObjeto():
  print('-------------------- Menu 3 de 3 - rota do objeto--------------------')
  while True:
    #Menu de seleção da rota
    print('Selecione a rota: \n',
          'BR - De Brasília para o rio de janeiro \n',
          'BS - De Brasília para São Paulo \n',
          'RB - Do Rio de Janeiro para Brasília \n',
          'RS - Do Rio de Janeiro para São Paulo \n',
          'SR - De São Paulo para o Rio de Janeiro \n',
          'SB - De São Paulo para Brasília')
    rotaObjeto = input('>>')
    #transformar o digito em maiúsculo
    rotaObjeto = rotaObjeto.upper()
    #retirar espaços vazios
    rotaObjeto = rotaObjeto.strip()

    if rotaObjeto == 'BR':
      return 1
    elif rotaObjeto == 'BS':
      return 1
    elif rotaObjeto == 'RB':
      return 1.2
    elif rotaObjeto == 'RS':
      return 1.2
    elif rotaObjeto == 'SR':
      return 1.5
    elif rotaObjeto == 'SB':
      return 1.5
    else:
      print('Você digitou uma rota que não existe!\n'
            'Por favor entre com rota desejada novamente')
      continue
#Fim da Função rotaObjeto

#Programa principal
print('-------------------- Bem Vindo a companhia de logistica Julio Henrique Passo de Araújo S.A! --------------------')
#Desempacotar o valores obtiodos pela função dimensoesObjeto, peso e rota.
valorObjeto, volumeObjeto = dimensoesObjeto()
print('O volume do objeto em cm³ é: {:.2f}'.format(volumeObjeto))
peso = pesoObjeto()
rotaRU4133412 = rotaObjeto()
#print do valor total e dos multiplicadores.
print('Total a pagar (R$): {} (dimensões: {} * peso: {} * rota: {})'.format(valorObjeto * peso * rotaRU4133412, valorObjeto, peso, rotaRU4133412))
```

```python
# Início das variáveis globais.
listaPeca = []
codigoPecaRU4133412 = 0
# Fim das variáveis globais.

# Início cadastrar peças.
def cadastrarPeca(codigo):
    print('Bem vindo ao menu cadastrar peças')
    print('O código da peça é: {}'.format(codigo))
    nome = input('Por favor digite o nome da peça: ')
    fabricante = input('Por favor digite o fabricante da peça: ')
    valor = int(input('Por favor digite o valor (R$) da peça: '))
    dicionarioPeca = {'código': codigo,     #Dicionário para receber os valores de código, nome, fabricante e valor
                      'nome': nome,
                      'fabricante': fabricante,
                      'valor': valor}
    listaPeca.append(dicionarioPeca.copy())
# Fim cadastrar peças.

# Início consultar peças.
def consultarPeca():
    print('Bem vindo ao menu consultar peças')
    while True:
        menuConsultar = input('\nEscolha a opção desejada: \n' +
                              '1 - Consultar todas as peças \n' +
                              '2 - Consultar peças por código \n' +
                              '3 - Consultar peças por fabricante \n' +
                              '4 - Retornar \n' +
                              '>>')
        if menuConsultar == '1':
            print('Você escolheu consultar todas as peças')
            for peca in listaPeca:  # peca vai varrer a listaPeca
                print('--------------------')
                for key, value in peca.items():  # vai varrer todos os conjuntos chave e valor do dicionário peca
                    print('{}: {}'.format(key, value))
                print('--------------------')

        elif menuConsultar == '2':
            print('Você escolheu consultar por código')
            valorDesejado = int(input('Entre com o código desejado: '))
            for peca in listaPeca:
                if peca['código'] == valorDesejado:  # O valor do campo cód é igual ao valor desejado.
                    print('--------------------')
                    for key, value in peca.items():  # vai varrer todos os conjuntos chave e valor do dicionário peca
                        print('{}: {}'.format(key, value))
                    print('--------------------')

        elif menuConsultar == '3':
            print('Você escolheu consultar peças por fabricante')
            valorDesejado = input('Entre com o fabricante desejado: ')
            for peca in listaPeca:
                if peca['fabricante'] == valorDesejado:  # O valor do campo cód é igual ao valor desejado.
                    print('--------------------')
                    for key, value in peca.items():  # vai varrer todos os conjuntos chave e valor do dicionário peca
                        print('{}: {}'.format(key, value))
                    print('--------------------')

        elif menuConsultar == '4':
            return  # Retorna ao menu anterior.

        else:
            print('Opção inválida. Tente novamente')
            continue  # volta para o inicio do laço.
# Fim consultar peças.

# Início remover peças.
def removerPeca():
    print('Bem vindo ao menu remover peças')
    valorDesejado = int(input('Entre com o código do produto que deseja remover: '))
    for peca in listaPeca:
        if peca['código'] == valorDesejado:  # O valor do campo cód é igual ao valor desejado.
            listaPeca.remove(peca)
            print('Peça removida')
# Fim remover peças.

# Início programa principal.
print('---Bem Vindo a controle de estoque da bicicletaria do  Julio Henrique Passo de Araújo!---')
while True:
    menuPrincipal = input('\nEscolha a opção desejada: \n' +
                          '1 - Cadastrar peças \n' +
                          '2 - Consultar peças \n' +
                          '3 - Remover peças \n' +
                          '4 - Sair \n' +
                          '>>')
    if menuPrincipal == '1':
        codigoPecaRU4133412 += 1
        cadastrarPeca(codigoPecaRU4133412)

    elif menuPrincipal == '2':
        consultarPeca()

    elif menuPrincipal == '3':
        removerPeca()

    elif menuPrincipal == '4':
        break  # Encerra o programa.

    else:
        print('Opção inválida. Tente novamente')
        continue  # volta para o inicio do laço.
# Fim programa principal.
```
